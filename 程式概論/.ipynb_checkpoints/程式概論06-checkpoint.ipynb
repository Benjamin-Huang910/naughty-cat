```python
#List常用函數
data3=[49,52,13,13,13,78,24,13]
print(data3)
#1
data3.append(100)#新增添加的元素(只可新增一個)
print('#1:',data3)
#2
data3.extend([14,5,2])#新增添加的串列
print('#2:',data3)
#3
data3.insert(0,55)#於第一個位置插入數字55作為data3的第一個項目
print('#3:',data3)
#4
data3.remove(52)#將data3裡52的這個數字從裏頭刪除(以元素值指定)
print('#4:',data3)
#5
ele5=data3.pop(5)#將串列與串列中的元素分離(以索引值指定)
print('#5:',ele5)
print(data3)
#6
data3.reverse()#將串列順序倒過來
print('#6:',data3)
#7
num13=data3.count(13)#去數串列中的特定元素在裡面有幾個
print('#7:',num13)
#8
i=data3.index(24)#找第一次出現的索引位置值
print('#8:',i)

i_13=data3.index(13)
print(i_13)
i_13=data3.index(13,i_13+1)#從指定的位置開始往後找
print(i_13)
#9
print('#9:',len(data3))#串列的總元素數目(長度)
#10
print('#10:',sum(data3))#串列的總和
#11
print('#11:',min(data3))#串列的最小值
#12
print('#12:',max(data3))#將data3裡最大的數字顯示出來
#13
data3_sorted=sorted(data3,reverse=True)#將data3裡的數字依大小順序由大到小排好
print('#13:',data3_sorted)
#14
data3.clear()#將data3裡的資料全數清空
print('#14:',data3)
```

```python
#字典的宣告
dict1={}
dict2=dict()
dict3={'Apple':50,'Orange':20,'Banana':15}
dict4={1.1:50,1.2:20,1.3:15}
dict5={1:50,2:20,3:15}
print(type(dict5))
print(dict3)
print(dict4)
print(dict5)

#字典的讀取
print(dict3['Orange'])
print(dict4[1.2])
print(dict5[2])

#增加新的字典元素
dict4['Lemon']=75
print(dict4)


#修改字典內的值
dict4['Lemon']=60
print(dict4)

#刪除字典內某個元素
del dict4['Lemon']
print(dict4)

#刪除整個字典
del dict5
#print(dict5) #error

#清空字典內的元素
dict4.clear()
print(dict4)

#字典常用的函數
print(len(dict3))

dict4=dict3.copy()
print(dict4)

print(dict4.get('Orange'))
print(dict4.get('Lemon'))
print(dict4.get('Lemon','找不到這個水果對應的值'))

print('Lemon' in dict4)
print('Apple' in dict4)

print(dict4.keys())

print(dict4.values())

print(dict4.items())

#字典中的Key如果重複了...
dict6={1:50,2:20,3:15,2:90}
print(dict6.get(2))

#運用字典來印出
#John is 12 years old
#Bob is 19 years old
#Cindy is 18 years old
name_age={"John":12,"Bob":19,"Cindy":18}
#第一種寫法
print("第一種寫法:")
print("for迴圈計數器的內容:",name_age.keys())
for name in name_age.keys():
    print(f'{name:s} is {name_age[name]:d} years old')
#第二種寫法
print("第二種寫法:")
print("for迴圈計數器的內容:",name_age.items())
for name,age in name_age.items():
    print(name,'is',age,'years old')

#第三種寫法
print("第三種寫法:")
print("for迴圈計數器的內容:",name_age)
for name in name_age:
    print('%s is %d years old'%(name, name_age[name]))
```

```python
#開啟一個檔案，把唐詩登鸛雀樓寫進檔案中。

#第一種寫法
file1=open('c:\\python\\files\\poem1.txt', 'a+',  encoding='utf-8')
file1.write('《登鸛雀樓》\n\n')
file1.write('白日依山盡，黃河入海流。\n')
file1.write('欲窮千里目，更上一層樓1。\n\n')
file1.close()  #用第一種寫法時，檔案讀寫完成後要加close()


#第二種寫法，檔案讀寫完成後，可以不用加close()
with open('poem2.txt', 'w+',  encoding='utf-8') as file2:
    file2.write('《登鸛雀樓》\n\n')
    file2.write('白日依山盡，黃河入海流。\n')
    file2.write('欲窮千里目，更上一層樓。\n\n')
#開啟一個檔案，把檔案中的資料都讀進來。
file1=open('poem2.txt', 'r', encoding='utf-8')
list_lines=file1.readlines() #把檔案內所有的內容讀進一個串列中
file1.close()
print('將檔案內容以串列的方式印出:', list_lines)  #以串列的型態印出來

print('\n將串列內的每個元素一個一個印出來:')
for line in list_lines:
    print(line, end='')   #串列內的元素一個一個印出來
```
