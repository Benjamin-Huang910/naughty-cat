```python
from datetime import *

#date類別
print("-"*10+"date"+"-"*10)
d1=date(2020,11,1)   #宣告一個date物件d1
print(d1)

print(d1.day) #印出屬性:天
print(d1.year) #印出屬性:年分
print(d1.month) #印出屬性:月份

print(d1.ctime())  #以字串回傳「星期  月  日 時:分:秒 年」
d1=d1.replace(2020,1,1)  #重設年月日
print(d1)  
print(d1.weekday())  #回傳星期幾，0代表星期一
print(d1.isoweekday())  #回傳星期幾，1代表星期一


d2=date.today()   #以今天的日期來宣告一個date物件
print(d2)

td2=d2.timetuple()   #以tuple方式回傳時間結構
print(td2)

diff=d1-d2         #兩個date物件相減，相減的結果是一個timedelta的物件diff
print(diff.days)   #取得物件diff的屬性days(天數)就是兩個日期間的差距


#time類別
print("-"*10+"time"+"-"*10)
t0=time()   #宣告一個time物件t0
print(t0)
t1=time(15,20,50)  #宣告一個time物件t1
print(t1)
t1=t1.replace(20,20,50)  #重設時間
print(t1)


#datetime類別
print("-"*10+"datetime"+"-"*10)
td=datetime.today()   #傳回當下的日期和時間
now=datetime.now()    #傳回當下的日期和時間
print(td)
print(now)
print(now.year)       #屬性year
print(now.month)      #屬性month
print(now.day)        #屬性day
print(now.hour)       #屬性hour
print(now.minute)     #屬性minute
print(now.second)     #屬性second
dt1=datetime.combine(d1,t1)   #將date物件和time物件組合成一個datetime物件
print(dt1)
df=dt1-now            #兩個datetime物件相減產生timedelta的物件
print(df.days)        #屬性days，將兩個日期間的差距以天數的方式顯示

print("-"*10+"others"+"-"*10)
print(type(dt1))
print(type(d1))
print(type(t1))
print(isinstance(dt1, datetime))
print(isinstance(d1, date))
print(isinstance(t1, time))
print(isinstance(df.days, int))
```

```python
import tkinter   


win=tkinter.Tk()   #建立新的主視窗，一個程式只能有一個主視窗
win.title('主視窗的標題')   #主視窗的標題
win.geometry('500x300')   #主視窗的大小，寬度x高度

secwin=tkinter.Toplevel(win,width=400, height=200)  #建立新的子視窗，可以有多個子視窗
secwin.title("子視窗的標題")    #子視窗的標題

win.mainloop()     #呼叫mainloop()函數讓程式運作直到主視窗被關閉為止
```

```python
win=tkinter.Tk()   #建立新視窗
win.title('視窗的標題')   #視窗的標題
win.geometry('500x300')   #視窗的大小，寬度x高度

#標籤
lab1=tkinter.Label(win, text="這是第一個標籤", bg="yellow")
lab2=tkinter.Label(win, text="這是第二個標籤", bg="blue")
lab3=tkinter.Label(win, text="這是第三個標籤", bg="red")
lab4=tkinter.Label(win, text="這是第四個標籤", bg="green", font=('新細明體',14),padx=40,pady=20)
#安排元件在視窗中的位子
##lab1.pack()   #無參數時，由系統決定  
##lab2.pack(side="bottom")  #預設值是 side="top"
##lab3.pack(side="right")
##lab4.pack(side="left")
#另一種安排元件在視窗中位子的方式，指定列、欄來放置元件
lab1.grid(row=0, column=0)
lab2.grid(row=0, column=2)
lab3.grid(row=1, column=1)
lab4.grid(row=2, column=2)

win.mainloop()     #呼叫mainloop()函數讓程式運作直到視窗被關閉為止
```

```python
def addition():
    result.set(num1.get() + num2.get()) #用get()來取得num1和num2的值
                                        #用set()來設定result的值
    
win = tkinter.Tk()
win.title('加法運算')

num1=tkinter.DoubleVar()  #DoubleVar(): 產生tkinter定義的浮點數型態物件
                          #IntVar():整數; StringVar():字串; BooleanVar():布林值 
num2=tkinter.DoubleVar()
result=tkinter.DoubleVar()

item1=tkinter.Entry(win, width=10, textvariable=num1)  #建立一個Entry元件
label1=tkinter.Label(win, width=5, text='+')
item2=tkinter.Entry(win, width=10, textvariable=num2)
btn=tkinter.Button(win, width=5, text='=', command=addition)
label2=tkinter.Label(win, width=10, textvariable=result)

item1.pack(side='left')
label1.pack(side='left')
item2.pack(side='left')
btn.pack(side='left')
label2.pack(side='left')

win.mainloop()
```
