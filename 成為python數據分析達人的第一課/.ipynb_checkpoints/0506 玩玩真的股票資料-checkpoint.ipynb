```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
import pandas_datareader as web
```

```python
df=web.DataReader("AAPL","yahoo",start="2014-9-1",end="2019-8-31")
```

```python
df.head()
```

```python
P=df["Adj Close"]
```

```python
P.head()
```

```python
P.plot()
```

# 計算報酬率

```python
P.diff()
```

```python
r=P.diff()/P
```

```python
r[-100:].plot()
```

# 移動平均

```python
P.rolling(window=20).mean()
```

```python
P.rolling(window=20).mean().plot()
```

```python
P.plot()
P.rolling(window=20).mean().plot()
P.rolling(window=60).mean().plot()
```
