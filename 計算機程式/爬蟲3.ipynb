```python
import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup
```

```python
def get_web_page(url):
    resp = requests.get(url)
    resp.encoding = 'utf-8'
    
    if resp.status_code != 200:
        print('Invalid url:', resp.url)
        return None
    else:
        return resp.text
```

```python
     if __name__ == "__main__":
        #第一段是針對我們所需要的網站區塊作的連結
        url='https://www.vgchartz.com/games/games.php?name=&keyword=&console=&region=All&developer=&publisher=&goty_year=&genre=&boxart=Both&banner=Both&ownership=Both&showmultiplat=No&results=100&order=CriticScore&showtotalsales=0&showpublisher=0&showpublisher=1&showvgchartzscore=0&showvgchartzscore=1&shownasales=0&showdeveloper=0&showcriticscore=0&showcriticscore=1&showpalsales=0&showreleasedate=0&showreleasedate=1&showuserscore=0&showuserscore=1&showjapansales=0&showlastupdate=0&showlastupdate=1&showothersales=0&showshipped=0&showshipped=1'
        rate_df = pd.read_html(url)[7] # 7 refers to the table index of this page
        rate_page = get_web_page(url)
        soup = BeautifulSoup(rate_page, 'html.parser')
        rate_list = soup.find_all('tr')[27:127] # table rows for storing the rates #這一行是爬遊戲評分的數據
        odd_trs = soup.find_all('tr', style='background-image:url(../imgs/chartBar_large.gif); height:70px')#這兩行是爬遊戲名稱的區塊
        even_trs = soup.find_all('tr', style='background-image:url(../imgs/chartBar_alt_large.gif); height:70px')
        
        #創建儲存資料的地方
        games_name= []
        VGChartz_Score = []
        Critic_Score = []
        User_Score = []
        
        #把爬到的數據丟進儲存資料
        for data in rate_list:
            row = data.find_all('td', align='center')
            VGChartz_Score.append(row[1].text.strip())
            Critic_Score.append(row[2].text.strip())
            User_Score.append(row[3].text.strip())
        
        #把爬到的遊戲名稱丟進儲存資料
        for i in range(len(odd_trs)):
            odd_game_name = odd_trs[i].find('td', style='font-size:12pt;').text.strip()
            even_game_name = even_trs[i].find('td', style='font-size:12pt;').text.strip()
    
            games_name.append(odd_game_name)
            games_name.append(even_game_name)     
        
        #把儲存資料轉換成dataframe
        rate_df = pd.DataFrame({
            'Game_Name': games_name,
            'VGChartz_Score': VGChartz_Score,
            'Critic_Score': Critic_Score,
            'User_Score': User_Score
        })
```

```python
#把dataframe輸出成excel檔
outputpath='遊戲評分_page2.xlsx'
rate_df.to_excel(outputpath,index=False,header=False)
```
