```python
import numpy as np
import pandas as pd

import requests
from bs4 import BeautifulSoup
```

```python
def get_web_page(url):
    resp = requests.get(url)
    resp.encoding = 'utf-8'
    
    if resp.status_code != 200:
        print('Invalid url:', resp.url)
        return None
    else:
        return resp.text
```

```python
if __name__ == "__main__":
    for i in range(1,3):
        url='https://www.vgchartz.com/games/games.php?page={}&results=50&name=&console=&keyword=&publisher=&genre=&order=CriticScore&ownership=Both&boxart=Both&banner=Both&showdeleted=&region=All&goty_year=&developer=&direction=DESC&showtotalsales=0&shownasales=0&showpalsales=0&showjapansales=0&showothersales=0&showpublisher=1&showdeveloper=0&showreleasedate=1&showlastupdate=1&showvgchartzscore=1&showcriticscore=1&showuserscore=1&showshipped=1&alphasort=&showmultiplat=No.html'.format(i) 
        rate_df = pd.read_html(url)[7] # 7 refers to the table index of this page
        rate_page = get_web_page(url)
        soup = BeautifulSoup(rate_page, 'html.parser')
        rate_list = soup.find_all('tr')[27:77] # table rows for storing the rates
        
        Game_Name= []
        VGChartz_Score = []
        Critic_Score = []
        User_Score = []

        for data in rate_list:
            row = data.find_all('td', align='center')
            title = data.find('a').string
            Game_Name.append(title)
            VGChartz_Score.append(row[1].text.strip())
            Critic_Score.append(row[2].text.strip())
            User_Score.append(row[3].text.strip())
   
    
        rate_df = pd.DataFrame({
            'Game_Name': Game_Name,
            'VGChartz_Score': VGChartz_Score,
            'Critic_Score': Critic_Score,
            'User_Score': User_Score
        })
```

```python
outputpath='遊戲評分.xlsx'
rate_df.to_excel(outputpath,index=False,header=False)
```
