```python
import matplotlib.pyplot as plt  # import matplotlib相關套件
from matplotlib.font_manager import FontProperties  # import字型管理套件
```

# 杜邦分析法

```python
year = [2016, 2017, 2018, 2019, 2020]  # 使用年份當做X軸資料
stock_2345 = [6.19, 6.64, 8.10, 8.75, 9.90]  # 使用智邦的ROEQ4當第一條線的資料
stock_3380 = [1.44, 0.81, 1.10, 0.55, 2.57]  # 使用明泰的ROEQ4當第二條線的資料
plt.figure(figsize=(15,10),dpi=100,linewidth = 2)  # 設定圖片大小為長15、寬10
plt.plot(year,stock_2345,'s-',color = 'r', label="2345")  # 把資料放進來並指定對應的X軸、Y軸的資料，用方形做標記(s-)，並指定線條顏色為紅色，使用label標記線條含意
plt.plot(year,stock_3380,'o-',color = 'g', label="3380")  # 把資料放進來並指定對應的X軸、Y軸的資料 用圓形做標記(o-)，並指定線條顏色為綠色、使用label標記線條含意
plt.title("ROE", size=40, x=0.5, y=1.03)  # 設定圖片標題，以及指定字型設定，x代表與圖案最左側的距離，y代表與圖片的距離
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)  # 设置刻度字体大小
plt.xlabel("year", fontsize=30, labelpad = 15)  # 標示x軸(labelpad代表與圖片的距離)
plt.ylabel("%", fontsize=30, labelpad = 20)  # 標示y軸(labelpad代表與圖片的距離)
plt.legend(loc = "best", fontsize=20)  # 顯示出線條標記位置
plt.show()  # 畫出圖片
```

```python
year = [2016, 2017, 2018, 2019, 2020]
stock_2345 = [0.52, 0.56, 0.61, 0.5, 0.53]
stock_3380 = [0.36, 0.32, 0.28, 0.24, 0.37]
plt.figure(figsize=(15,10),dpi=100,linewidth = 2)
plt.plot(year,stock_2345,'s-',color = 'r', label="2345")
plt.plot(year,stock_3380,'o-',color = 'g', label="3380")
plt.title("Asset Turnover", size=40, x=0.5, y=1.03)
plt.xticks(fontsize=20)
plt.yticks(fontsize=20)
plt.xlabel("year", fontsize=30, labelpad = 15)
plt.ylabel("rate", fontsize=30, labelpad = 20)
plt.legend(loc = "best", fontsize=20)
plt.show()
```
