## 問題1
* 問題：如果自己是推銷保險的人，如何知道自己能夠推銷出去的成功率？
* 輸入：被推銷的人的表情
* 輸出：被推銷成功的人的表情設1，推銷失敗的人的表情設0
* 訓練資料：拍照前要先跟被拍的人申明這是一個實驗，不會去侵犯他的個資，然後拍他的正面照；可多組人同時進行

## 問題2
* 問題：判斷貓咪是否要上廁所(為了避免上在貓砂外)
* 輸入：貓咪日常的各種表情照片
* 輸出：貓咪在上廁所的表情照片
* 訓練資料：自己拍，或是到網路去搜集其他人的照片
* problem:資料不好蒐集

## 問題3
* 問題：能否有一個裝置是可以偵測一個人的酒精濃度超標，並發出警告聲制止其開車
* 輸入：人的日常表情圖片
* 輸出：人酒醉的表情圖片
* 訓練資料：尋找網路上的照片，或是去新聞搜尋照片

## 問題4
* 問題：藉由30秒的旋律判斷這是哪一首歌
* 輸入：一首歌裡隨機30秒片段的旋律
* 輸出：一首完整的歌
* 訓練資料：從youtube抓歌
* problem:與搜尋系統差別甚小

## 問題5
* 問題:市面上有許多不同價格的水，撕掉包裝的話，能夠分辨出哪一個是哪種水嗎
* 輸入:蒐集各個品牌的水，然後同樣都倒到玻璃杯裡面
* 輸出:為每種品牌的水設立代號或數字(如:A、B、C、1、2、3...)
* 訓練資料:分組進行，設定每一天拍一百張水的照片，可重複，但從不同的角度拍攝

## 問題6
* 問題:如果今天上市了一個新遊戲，還沒有相關的評價，可不可以由程式來評斷其遊玩分數?
* 輸入:家用主機、掌上型主機、電腦的遊戲基本劇情架構、遊玩類型(例:開放世界、橫向卷軸等)、畫質
* 輸出:遊戲的評分劃分為兩種，如果分數為1~10分，則5分以下為0，5分以上為1
* 訓練資料:去搜尋目前還存在的遊戲公司的資料庫，尋找他們近五年的遊戲然後輸入評分

## 問題7
* 問題:診所大部分的病人都是感冒類的病，能否直接由電腦判斷他是不是感冒然後給他開出配方，節省過多重複的動作
* 輸入:來看診的病人的臉
* 輸出:症狀為感冒的病人的臉設參數為1,不是的為0
* 訓練資料:通常診所裡會有監視器，藉由監視器的畫面去蒐集病人的五官(一般的診所一天為100~300人不等)
